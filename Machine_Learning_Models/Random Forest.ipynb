```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
training_df = pd.read_csv('../train.csv')
training_df.set_index('Timestamp', inplace=True)
training_df.drop(['Country', 'Ad Topic Line'], axis=1, inplace=True)
training_df.head()
```

```python
X = training_df.loc[:, 'Daily Time Spent on Site': 'Ad Line Word Length']
y = training_df['Clicked on Ad']
```

```python
rfc = RandomForestClassifier(random_state=42)
```

```python
param_grid = [
    {
        'max_features': ['auto', 'sqrt'], 
        'n_estimators': [100, 250, 500, 1000],
        'min_impurity_decrease': [0.0, 1e-7]
    }
]
```

```python
cv_len = 5

rgc_grid_search = GridSearchCV(rfc, param_grid, cv=cv_len, scoring='accuracy', refit='accuracy')
```

```python
rgc_grid_search.fit(X, y)
```

```python
# rgc_grid_search.cv_results_
```

```python
rgc_grid_search.best_estimator_
```

```python
for i in range(cv_len):
    for score, param in zip(rgc_grid_search.cv_results_[f'split{i}_test_score'], rgc_grid_search.cv_results_['params']):
        print(f'{param}: {score}')
```
