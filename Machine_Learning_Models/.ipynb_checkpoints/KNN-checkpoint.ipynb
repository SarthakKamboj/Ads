```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
import pickle
```

```python
training_df = pd.read_csv("../train.csv")
```

```python
training_df.set_index("Timestamp",inplace=True)
training_df.head()
```

```python
training_X_cols = list(training_df.select_dtypes(exclude=["object"]).columns)
training_X_cols.remove("Clicked on Ad")
training_X_cols
```

```python
training_y_col = "Clicked on Ad"
```

```python
training_X = training_df[training_X_cols]
training_y = training_df[training_y_col]
```

```python
param_grid = [
    {"n_neighbors":[2,4,6,8,10]}
]
```

```python
grid_search = GridSearchCV(KNeighborsClassifier(),param_grid,cv=5,scoring="accuracy")
```

```python
grid_search.fit(training_X,training_y)
```

```python
grid_search.best_estimator_
```

```python
knn_file = open("Best_Models/KNN.pkl","wb")
pickle.dump(grid_search.best_estimator_,knn_file)
```

```python
results = grid_search.cv_results_
for key in results.keys():
    print(key)
    print(results[key])
    print("\n\n")
```

```python
for mean_score, params in zip(results["mean_test_score"], results["params"]):
    print(f"mean score: {mean_score} params: {params}")
```
