```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
```

```python
training_df = pd.read_csv("../train.csv")
```

```python
training_df.set_index("Timestamp", inplace=True)
training_df.head()
```

```python
training_X_cols = list(training_df.select_dtypes(exclude=["object"]).columns)
training_X_cols.remove("Clicked on Ad")
training_X_cols
```

```python
training_y_col = "Clicked on Ad"
```

```python
training_X = training_df[training_X_cols]
training_y = training_df[training_y_col]
```

```python
param_grid = [
    {"loss":["hinge","log"], "alpha":[0.01,0.001,0.0001]}
]
```

```python
grid_search = GridSearchCV(SGDClassifier(),param_grid,cv=5,scoring="accuracy")
```

```python
grid_search.fit(training_X,training_y)
```

```python
best_estimator = grid_search.best_estimator_
print(best_estimator)
```

```python
results = grid_search.cv_results_
for key in results.keys():
    print(key)
    print(results[key])
    print("\n\n")
```

```python
for accuracy, params in zip(results["mean_test_score"], results["params"]):
    print(f"params: {params} accuracy: {accuracy} ")
    print("\n")
```
